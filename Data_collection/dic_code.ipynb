```python
import re
import pandas as pd
```

```python
# import the dictionary 
dic=pd.read_csv("yearcode.csv",index_col=0, squeeze=True).to_dict()
dic2=pd.read_csv("yearcode2.csv",index_col=0, squeeze=True).to_dict()
#dic
```

```python
## import the metadata
## here metadata is a tab delimited file
meta=pd.read_csv("RSVB_meta.txt",sep='\t')
```

```python
## do translate
meta['yearcode']=meta['year'].map(dic2)
meta
```

```python
## combine region and yearcode
meta['code'] = meta['yearcode'].str.cat(meta['region'], sep =".")
meta
## count how many group in code
meta.code.drop_duplicates().count()
#meta.code.drop_duplicates()
```

```python
## save dataframe as tab.txt
meta.to_csv('RSVB_metacode.txt',sep='\t', index=False)
```
