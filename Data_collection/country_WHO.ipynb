```python
# Creating a Dictionary
dic = {}
with open("country_region.txt", 'r') as f:
    for line in f:
        (x,y) = line.split(',')
        dic[x] = y
       # print(dic)
        #print(len(dic))
        
#use dictionary to code input data
outfile=open("code_output.txt",'w')

with open("country.txt", 'r') as f1:
    for line in f1:
        x = line.strip('\n')
        if x in dic:
            print(x +'\t'+ dic[x])
            outfile.write(x +'\t'+ dic[x]+'\n')
        if x not in dic:
            print(x+'\t'+"na"+"\n")
            outfile.write(x+'\t'+"na"+'\n')



outfile.close()
```
